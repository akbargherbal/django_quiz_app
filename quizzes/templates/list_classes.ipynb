```python
import pandas as pd
import regex as re
import os
```

```python
from scrapy import Selector
def get_all_classes(text):
    result = Selector(text=str(text)).xpath('//*[@class]//@class').getall()
    result = [x.split() for x in result]
    result = [item for sublist in result for item in sublist]
    return result

list_txt = []
for root, dirs, files in os.walk('./quizzes'):
    for file in files:
        if file.endswith('.html'):
            list_txt.append(os.path.join(root, file).replace(os.sep, '/'))

list_classes = []

for file in list_txt:
    with open(file, 'r', encoding='utf-8') as f:
        text = f.read()
        list_classes.extend(get_all_classes(text))

set_classes = set(list_classes)
```
